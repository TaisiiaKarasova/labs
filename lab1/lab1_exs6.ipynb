```python
# Exercise 3

import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-30,30,0.01)
plt.plot(x, (np.log(np.power(x, 2) + 1) / np.log(1+np.tan(1 / (1 + np.power(np.sin(x), 2)))))*(np.power(np.e, -(abs(x) / 10))), 'b')
plt.title(r'$f(x)=log_{(1+\tan(\frac{1}{\sin(x)^2+1}))}(x^2+1)\exp(-\frac{|x|}{10})$', size=20, color='red')
plt.xlabel(r'$x$', size=15)
plt.ylabel(r'$f(x)$', size=15)
plt.grid(True)
plt.savefig('fig_lab1_ex3.png')
plt.show()
```

```python
# Exercise 5

import numpy as np
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5, 6]
y = [1, 1.42, 1.76, 2, 2.24, 2.5]
p, v = np.polyfit(x, y, deg=2, cov=True)
p_f = np.poly1d(p)
l, k = np.polyfit(x, y, deg=1, cov=True)
l_f = np.poly1d(l)

# subplot 1
sp = plt.subplot(121)
plt.errorbar(x, y, xerr=0, yerr=y-p_f(x), alpha=0.6)
plt.plot(x, p_f(x))
plt.grid(True)
plt.title(r'$deg=2$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.axis([0.8, 6.2, 0.8, 2.7])
plt.show()

# subplot 2
sp = plt.subplot(122)
plt.errorbar(x, y, xerr=0, yerr=y-l_f(x), alpha=0.6)
plt.plot(x, l_f(x))
plt.grid(True)
plt.title(r'$deg=1$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.axis([0.8, 6.2, 0.8, 2.7])
plt.show()

```

```python
# Exercise 6

import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-2, 2, 0.01)
y = 0
for n in range(0, 500, 1):
    y += 0.5**n * np.cos(3**n * np.pi * x)
plt.plot(x, y, 'r')
plt.xlabel(r'$x$', fontsize = 15)
plt.ylabel(r'$f(x)$', fontsize = 15)
plt.savefig('fig_lab1_ex6.png')
plt.show()
```
